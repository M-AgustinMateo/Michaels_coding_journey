```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
data = pd.read_csv("AAPL.csv", index_col=0, parse_dates=True)
```

```python
data["%-chg"] = data['Close'].pct_change()
```

```python
data.head()
```

```python
import numpy as np
```

```python
data['Log returns']=np.log(data['Close']/data["Close"].shift())
```

```python
data.head()
```

```python
data['Log returns'].std()
```

```python
volatility = data['Log returns'].std()*252**0.5
```

```python
volatility
```

```python
fig, ax = plt.subplots()
data['Log returns'].hist(ax=ax, bins=50, alpha=0.6, color='r')
ax.set_xlabel
```

```python
data['MA10']=data['Close'].rolling(10).mean()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
data = pd.read_csv("AAPL.csv", index_col=0, parse_dates=True)
```

```python
data.head()
```

```python
exp1 = data['Close'].ewm(span=12, adjust=False).mean()
exp2 = data['Close'].ewm(span=26, adjust=False).mean()
data['MACD'] = exp1 - exp2
data['Signal line'] = data['MACD'].ewm(span=9, adjust=False).mean()
```

```python
data.tail()
```

```python
fig, ax =plt.subplots()
data[['MACD','Signal line']].plot(ax=ax)
data['Close'].plot(ax=ax, alpha = 0.70, secondary_y=True)
```
